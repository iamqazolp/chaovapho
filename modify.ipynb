```python
import pandas as pd
import numpy as np
```

```python
def create_date(start,end,month):
    return [f'{x}/{month}' for x in range(start,end+1)]
```

```python
ds_hocsinh=['Thái','Linh','Đức','Ngọc','Cương','Chinh','Long','Hải','Phu','Hiếu','Minh','Hiệp','Khôi',"Huy"]
ds=pd.DataFrame(data=np.zeros(shape=(14,30),dtype=int),columns=create_date(2,31,8),index=ds_hocsinh)
```

```python
def change_values(date:list,name:list,val):
    global ds
    for i in date:
        for j in name:
            ds[i][j]=val
    ds['Tổng']=ds[list(ds.columns[:ds.shape[1]-1])].sum(axis=1)
```

```python
change_values(['2/8'],['Ngọc','Hiếu'],2)
change_values(['3/8'],['Ngọc','Cương'],2)
change_values(['5/8'],['Minh','Hiệp','Khôi'],2)
change_values(['6/8'],['Long'],2)
change_values(['7/8'],["Chinh",'Cương','Ngọc'],2)
change_values(['9/8'],['Ngọc'],2)
change_values(['9/8'],['Hiếu'],2)
change_values(['10/8'],['Ngọc','Cương','Huy'],2)
change_values(['11/8','15/8'],['Thái','Linh','Đức'],2)
change_values(['14/8'],['Cương','Chinh','Huy'],2)
change_values(['12/8'],['Phu','Hải'],1)
change_values(['14/8'],['Hiếu'],1)
change_values(['16/8'],['Ngọc'],2)
change_values(['17/8'],['Long','Chinh','Huy','Cương','Hiếu'],2)
change_values(['21/8'],['Cương','Ngọc','Huy','Chinh'],2)
change_values(['18/8'],['Hiệp','Minh'],1)
change_values(['26/8'],['Hiệp','Minh'],2)
change_values(['18/8','25/8'],['Hải','Phu'],2)
change_values(['23/8'],['Hiếu'],2)
change_values(['27/8'],['Long'],2)
change_values(['28/8'],['Cương','Ngọc','Huy','Chinh'],2)
change_values(['30/8'],['Ngọc'],1)

```
